### Exercise 1
(4 steps)

| Account holder | Account number | Balance |
| --- | --- | --- |
| Kenneth Effe | 3421 1242 3232 5324 | 232.2 |
| James Chang | 7583 0274 7537 9613 | 405.37 |
| Ralph Mallets | 8534 7319 4697 1254 | 342.65 |

````python
#step 1
data = {
    'Account holder': ['Kenneth Effe', 'James Chang', 'Ralph Mallets'],
    'Account number': ['3421 1242 3232 5324', '7583 0274 7537 9613', '8534 7319 4697 1254'],
    'Balance': [232.2, 405.37, 342.65]
}

bankaccounts = pd.DataFrame(data)

#step 2: set first column as index
bankaccounts = bankaccounts.set_index('Account holder')

#step 3: locate Ralph's account number and balance.
located_account = bankaccounts.loc['Ralph Mallets']

print(located_account)

print()

#step 4: calculate sum and average of the balances
sum = bankaccounts['Balance'].sum()
avg = bankaccounts['Balance'].mean()
print('sum:',sum)
print('average:',avg)
````
